# Word Graph - Triple Generation

An issue with the word graph + concept mapping is that some words may be assigned to multiple concepts. So it doesn't work well in it's current configuration.

Attr: https://stackoverflow.com/questions/12718537/generating-a-directed-word-graph-in-python

Schema for word graphs. Each work order entry will have its own sub-graph. These can then be filtered and aggregated.<br>

{"data": [{"uid": "", "assetid": "", "text": "", "creationdate": "", "concept": "", "tokenized": [], "nodes": [], "links": []}]}

For power BI implementation the data will be made into a CSV with a primary key and source/target columns of words in the word graphs.

Each work order has one word graph associated to it at an instance in time

## Construct data

```python
dfMWO = pd.read_csv('../data/ALOCA - pump-data.csv', engine='python')
```

```python
graphData = {"data": []}
conceptList = []

for index, row in tqdm(dfMWO.iterrows()):
    
    uniqueID = uuid.uuid1().int
    text = row["WO_DESCRIPTION"]
    creationDate = row["CREATION_DATE"].strftime("%d-%m-%Y")
    try:
        concept = row["ACTIVITY_TYPE"].lower()
    except:    # concepts maybe 'null' which can't be picked up via if-else statements.
        concept = 'unknown'
    tokenizedText = word_tokenize(row["WO_DESCRIPTION"].strip().lower())
    assetId = row["ASSET_NUMBER"]
    
    graphData["data"].append({"uid": uniqueID,
                               "assetid": assetId,
                               "text": text,
                               "creationdate": creationDate,
                               "concept": concept,
                               "tokenized": tokenizedText})
    
    if row["ACTIVITY_TYPE"] not in conceptList:
        conceptList.append(row["ACTIVITY_TYPE"])
```

```python
conceptString = "Concepts Identified:\n\t"+"\n\t".join([str(concept) for concept in conceptList])
```

```python
print(conceptString)
```

```python
# Generate sub-word graphs for each MWO instance and append to graph data dictionary.
graphDataIndex = 0

for doc in tqdm(graphData["data"]):
#     print(doc)
       
    # initialise directed graph
    dGTemp = nx.DiGraph()
    
    text = doc["tokenized"]
    concept = doc["concept"]
    uid = str(doc["uid"])
    
    # populate graph
    for i, word in enumerate(text):
        try:
            nextWord = text[i+1]
#             print(f'first word: {word} - next word: {nextWord}')
            
            # If the graph doesn't have a word node - add it.
            if not dGTemp.has_node(word):
                dGTemp.add_node(word)
                dGTemp.node[word]['count'] = 1
                dGTemp.node[word]['concept'] = concept
                dGTemp.node[word]['uid'] = uid

            # If the word is already in the graph - add a count to it.
            else:
                dGTemp.node[word]['count'] += 1
                dGTemp.node[word]['concept'] = concept
                dGTemp.node[word]['uid'] = uid

            if not dGTemp.has_node(nextWord):
                dGTemp.add_node(nextWord)
                dGTemp.node[nextWord]['count'] = 1
                dGTemp.node[word]['concept'] = concept
                dGTemp.node[word]['uid'] = uid
                
            else:
                dGTemp.node[nextWord]['count'] += 1
                dGTemp.node[word]['Concept'] = concept
                dGTemp.node[word]['uid'] = uid

            if not dGTemp.has_edge(word, nextWord):
                dGTemp.add_edge(word, nextWord, weight = 1, frequency = 1)
    
            else:
                dGTemp.adj[word][nextWord]['weight'] += 1
                dGTemp.adj[word][nextWord]['frequency'] += 1

        except IndexError:
            # Last word; no 'nextWord'
#             print('In exception')

            if not dGTemp.has_node(word):
                dGTemp.add_node(word)
                dGTemp.node[word]['count'] = 1
                dGTemp.node[word]['concept'] = concept
                dGTemp.node[word]['uid'] = uid
#                 print('left exception')
            else:
                dGTemp.node[word]['count'] += 1
                dGTemp.node[word]['concept'] = concept
                dGTemp.node[word]['uid'] = uid
#                 print('left exception')
                
        except:
            raise
     
    # convert graph to JSON and add nodes/links to mwo document
    dGTempJSON = json_graph.node_link_data(dGTemp)
    
    graphData["data"][graphDataIndex]["nodes"] = dGTempJSON["nodes"]
    graphData["data"][graphDataIndex]["links"] = dGTempJSON["links"]
    
    graphDataIndex +=1
```
