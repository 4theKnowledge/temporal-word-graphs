# Basic Word Graph
This notebook was originally developed for the example in section 7.6.1. Multi-sentence fusion via graphs in the textbook 'Automatic Text Summarization' by Moreno. It has now been updated to be used for maintenance work order presentation.

An issue with the word graph + concept mapping is that some words may be assigned to multiple concepts. So it doesn't work well in it's current configuration.

Attr: https://stackoverflow.com/questions/12718537/generating-a-directed-word-graph-in-python

```python
import networkx as nx
from sys import maxsize
import pandas as pd
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 12, 10
```

```python
# Generate corpus using one asset
# Schema will be {"Asset_Number": [Text": "", "Concept": "", "Year": "", "Month": ""]}


corpus = {assetNumber: []}
conceptList = []

for index, row in dfMWOSubset.iterrows():
    
    corpus[assetNumber].append({"Text": row["WO_DESCRIPTION"],"Concept": row["ACTIVITY_TYPE"],"Year": row["CREATION_DATE"].year, "Month": row["CREATION_DATE"].month})
    
    if row["ACTIVITY_TYPE"] not in conceptList:
        conceptList.append(row["ACTIVITY_TYPE"])
```

```python
corpus[assetNumber][:10]
```

```python
conceptList
```

```python
# preprocess text
# outputs tokenized sentence with start and end tokens
startToken = "<s> "
finishToken = " <f>"

corpusPreprocessed = {assetNumber: []}

for sentence in corpus[assetNumber]:
    sentenceDict = sentence
    sentence = sentence["Text"]
    sentence = startToken + sentence
    sentence = sentence.lower()
#     sentence = sentence.replace('in the', 'in_the')
#     sentence = sentence.replace('16th', 'sixteenth')
    sentence = sentence.replace('.', '')
    sentence = sentence.replace(',', '')
#     sentence = sentence.replace('1500s', '1500')
    sentence = sentence + finishToken
    
    # tokenize sentence
    sentence = sentence.split(' ')

    sentenceDict['P-Tokenized'] = sentence
    corpusPreprocessed[assetNumber].append(sentenceDict)
```

```python
corpusPreprocessed
```

```python
# Initialise directed graph
dG = nx.DiGraph()

# Construct directed graph
for doc in corpusPreprocessed[assetNumber]:
    sentence = doc['P-Tokenized']
    concept = doc['Concept']
    
    for i, word in enumerate(sentence):

        try:
            nextWord = sentence[i + 1]
            
#             print(f'first word: {word} - next word: {nextWord}')
            
            # If the graph doesn't have a word node - add it.
            if not dG.has_node(word):
                dG.add_node(word)
                dG.node[word]['count'] = 1
                dG.node[word]['Concept'] = concept

            # If the word is already in the graph - add a count to it.
            else:
                dG.node[word]['count'] += 1
                dG.node[word]['Concept'] = concept

            if not dG.has_node(nextWord):
                dG.add_node(nextWord)
                dG.node[nextWord]['count'] = 1
                dG.node[word]['Concept'] = concept
                
            else:
                dG.node[nextWord]['count'] += 1
                dG.node[word]['Concept'] = concept

            if not dG.has_edge(word, nextWord):
                dG.add_edge(word, nextWord, weight = 1, frequency = 1)
    
            else:
                dG.adj[word][nextWord]['weight'] += 1
                dG.adj[word][nextWord]['frequency'] += 1

        except IndexError:
            # Last word; no 'nextWord'
#             print('In exception')

            if not dG.has_node(word):
                dG.add_node(word)
                dG.node[word]['count'] = 1
                dG.node[word]['Concept'] = concept
#                 print('left exception')
            else:
                dG.node[word]['count'] += 1
                dG.node[word]['Concept'] = concept
#                 print('left exception')
                
        except:
            raise
```

```python
# Review node concepts and remove concepts for special tokens
for node in dG.nodes():
    if (node == '<s>') or (node == '<f>'):
        dG.node[node]["Concept"] = 'Special Token'
    print(f'Node: {node} | Concept: {dG.node[node]["Concept"]}')
```

```python
# Node frequency attributes
# Duplicates count for some reason...
for node in dG.nodes():
    print(f'Node: {node} | Count: {dG.node[node]["count"]}')
```

```python
# Generate cmap and make concept-hex mapping
cmap = plt.cm.get_cmap('seismic', len(conceptList))
```

```python
colorMap = {}

for concept in conceptList:
    rgb = cmap(conceptList.index(concept))[:3]
    
    colorMap[concept] = matplotlib.colors.rgb2hex(rgb)

# Add special token colour - these don't have concepts.
colorMap["Special Token"] = "#636363"
```

```python
colorMap
```

```python
# Assign node colours based on their concept property
graph_color_map = []
print('Colouring nodes')
for node in dG.nodes():
    graph_color_map.append(colorMap[dG.node[node]["Concept"]])
```

```python
# Get maximum edge weight to normalise edges against.
# attr: https://stackoverflow.com/questions/52440518/finding-maximum-weighted-edge-in-a-networkx-graph-in-python
maxEdgeWeight = max(dict(dG.edges).items(), key=lambda x: x[1]['weight'])[1]['weight']
scaleFactor = 4  # adds a bit more thickness to the lines.

# Normalize edge weights based on maxEdgeWeight
for edge in dG.edges():    
    dG.adj[edge[0]][edge[1]]["weight"] = dG.adj[edge[0]][edge[1]]["weight"]/maxEdgeWeight
#     print(f'Edge: {edge} | Weight: {dG.adj[edge[0]][edge[1]]["weight"]}')
```

```python
# Review edge frequencies
for edge in dG.edges():    
    print(f'Edge: {edge} | Frequency: {dG.adj[edge[0]][edge[1]]["frequency"]}')
```

```python
pos=nx.spring_layout(dG)
```

```python
# Get edge labels
edge_labels = nx.get_edge_attributes(dG, 'frequency')
```

```python
nx.draw(dG,
        with_labels=True,
        pos=pos,
        k=1,
        iterations=50,
        node_color=graph_color_map,
        width=[d['weight'] for _, _, d in dG.edges(data=True)],
        font_size=12)

nx.draw_networkx_edge_labels(dG,
                             pos,
                             edge_labels=edge_labels,
                             label_pos=0.5,
                             font_size=12)

# dummy scatter plot to show node labels
for label, color in colorMap.items():
    plt.scatter([],[], c=color, label=label)

plt.axis('off')
plt.legend(prop={'size': 12}, frameon=True, title="Concept", title_fontsize=14, facecolor='#f0f0f0')
plt.show()
```

## Basic Graph Analytics

```python
# Get all simple paths within directed graph
all_simple_paths = list(nx.all_simple_paths(dG, source='<s>', target='<f>'))
```

```python
# Filter paths for those that visit nodes of interest
# This will return any path where any of the nodes of interest are mathced. Not the set of them.
interestNodes = ['review', 'perform']
paths_with_key_nodes = [path for path in all_simple_paths if any(elem in interestNodes for elem in path)]
```
